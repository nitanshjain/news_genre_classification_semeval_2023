```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/nitanshjain/Documents/Projects/Sem_Eval/submission_albert_1.csv')
```

```python
df.head()
```

```python
x = df.to_string(header=False,
                  index=False,
                  index_names=False).split('\n')
vals = [','.join(ele.split()) for ele in x]
print(vals[0])
```

```python
new_vals = []
for val in vals:
    # replacing ,0 with ,
    val = val.replace(',0', '')
    new_vals.append(val)
```

```python
new_vals
```

```python
# replacing first comma with tab space
final_vals = []
for val in new_vals:
    val = val.replace(',', '\t', 1)
    final_vals.append(val)
```

```python
final_vals
```

```python
file = open('/Users/nitanshjain/Documents/Projects/Sem_Eval/submission.txt','w')
for item in final_vals:
	file.write(item+"\n")
file.close()
```
